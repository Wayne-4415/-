```python
import math 
import numpy as np
from math import exp, pi, sqrt, ceil, sin, cos
from numpy import linalg as LA
import time

import matplotlib as mpl
import matplotlib.pyplot as plt
#全域變數
N = 17
L = 2.6
def K(x,y,xc,yc):
     #參數未定 N A M lamb delta_z
    #N =
    #A = 
    k = 10 #temp
    delta_z = 1 
    #化為弦函數
    a = complex(cos(k * delta_z * (x**2 + y**2)), sin(k * delta_z * (x**2 + y**2)))
    if sqrt((x - xc)**2 + (y - yc)**2) < 1.2:
        return np.array(a) 
    else:
        return np.array(0)
def J(x,y,S):
    if sqrt(x**2 + y**2) < S:
        return 1/pi/S**2
    else:
        return 0
def T(f1,f2,g1,g2):  
    result = 0
    for i in range(N):
        for j in range(N):
            temp = K(f1, g1, i*L/N - 0.5*L, j*L/N - 0.5*L)
            temp2 = K(f2, g2, i*L/N - 0.5*L, j*L/N - 0.5*L)
            result += J(i*L/N - 0.5*L, j*L/N - 0.5*L, 0.5)*temp*np.conjugate(temp2)
    return result

def TT():
    TT = np.zeros((N**2,N**2))
    for i in range(N**2):
        for j in range(N**2):
            TT[i][j] = abs(T(((i+1) - N*(ceil((i+1)/N) - 1))*L/N - 0.5*L - L/2/N,
                             ((j+1) - N*(ceil((j+1)/N) - 1))*L/N - 0.5*L - L/2/N,
                             (ceil((i+1)/N))*L/N - 0.5*L - L/2/N,
                             (ceil((j+1)/N)*L/N - 0.5*L - L/2/N)
                            ))           
    return TT
        
def main():
    
    # 開始測量
    start = time.time()

    # 要測量的程式碼
    array = TT()
    #w, v = LA.eig(array) #w eigenvalue V eigenvector
    fig, ax = plt.subplots()
    ax.contour(array)
    plt.show()

    # 結束測量
    end = time.time()

    # 輸出結果
    print("執行時間：%f 秒" % (end - start))


main()
```
