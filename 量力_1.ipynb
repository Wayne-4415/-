```python
import numpy as np
from numpy import pi, sqrt, arctan
import matplotlib as mpl

v = np.arange(0, 10, 0.01)
b = np.arange(0,1,0.001)
eq = lambda b,r,m: ((m*pi/2 + arctan(sqrt(b/(1-b)))/2 + arctan(sqrt((b+r)/(1-b)))/2 )/sqrt(1-b))
```

```python
X, Y = {}, {}

for m in range(0,5,1):    
    x1, y1 = [], []
    for i in range(0,len(v),1):
        EQ = eq(b,0,m)
        eq2 = EQ - v[i]
        for j in range(0,len(eq2)-1,1):
            if(eq2[j]*eq2[j+1])<= 0:
                y1.append((b[j] + b[j+1])/2)
                x1.append(v[i])
    X.update({m:x1})
    Y.update({m:y1})

import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 80
mpl.rcParams['figure.figsize'] = [10, 10]  #[x,y] 
mpl.rcParams['font.size'] = 13

for i in range(0, len(X),1):
    plt.plot(X[i],Y[i])
plt.xlabel(r'$\nu$')
plt.ylabel('b')
plt.show()

```
