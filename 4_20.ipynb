```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
import math
# 上課ppt
#numerical integrations  func interpolation 2021  PAGE 18~END
# Partial differential equations (PDEs) 2021
X0 = 0
X1 = 1 / 6
X2 = 1 / 3
f0 = 1
f1 = 3**0.5/2 
f2 = 0.5
def f(x):
    answer = ((x - X1) * (x - X2)/ (X0 - X1) /(X0 - X2) ) * f0 + \
    ((x - X0) * (x - X2)/ (X1 - X0) /(X1 - X2) ) * f1 + \
    ((x - X0) * (x - X1)/ (X2 - X0) /(X2 - X1) ) * f2
    return answer

fig, ax = plt.subplots()
xx = np.array([0,1/6,1/3])
y = np.array([1,3**0.5/2,0.5])
x = np.linspace(0,0.5,1000)
ax.plot(x,f(x),label = "2nd order lanrange") 
ax.plot(xx,y,'o',label = "data")
ax.legend()
```

```python
N = 11
L = 1
h = L / (N-1)
X = np.arange(0,N)
X = h*X
V = np.zeros(N)
rho = np.zeros(N)
rho[int(N/2)] = 1
A = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        if i==j+1:
            A[i,j] = 1
        elif j==i+1:
            A[i,j] = 1
        elif i==j:
            A[i,j] = -2
## BC
A[0,0] = 1
A[0,1] = 0
A[N-1,N-1] = 1
A[N-1,N-2] = 0
print(A)
B = np.linalg.inv(A)
V = -h * np.matmul(B,rho)
fig, ax = plt.subplots()

ax.plot(X,V)
```
