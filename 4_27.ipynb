```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
import math
# 上課ppt
#Partial differential equations (PDEs)  page 18~

#Electric field induced by a point charge
delta = 0.001
def f(x,y):
    return 1 /((x**2 + y**2)**0.5 + 0.001)
def dfx(x,y):
    return x /((x**2 + y**2)**0.5 + 0.001)
def dfy(x,y):
    return y /((x**2 + y**2)**0.5 + 0.001)

#np.seterr(all='ignore') 

nx, ny = (11, 11)
x = np.linspace(-1, 1, nx)
y = np.linspace(-1, 1, ny)
xv, yv = np.meshgrid(x, y)

fig, ax = plt.subplots()
ax.contourf(xv, yv,f(xv,yv),50)
ax.quiver(xv, yv, dfx(xv,yv) , dfy(xv,yv))
plt.show()
#ax.contourf(xv, yv, f)
```
